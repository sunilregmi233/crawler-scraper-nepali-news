```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties


font_path = '/home/sunil/Documents/Assignment_IS/crawler-scraper-nepali-news/Preeti/mangal.ttf'
font = FontProperties(fname=font_path)

plt.rcParams["figure.figsize"] = (20,10)

keys = twenty_words.keys()
print(keys)



fig,ax = plt.subplots()
plt.bar(twenty_words.keys(), twenty_words.values(), align='center')
for label in ax.get_xticklabels():
    label.set_fontproperties(font)
plt.title('Term Frequency Graph')
plt.xlabel("Word")
plt.ylabel("Count")
plt.xticks(rotation=90, fontsize=14)
plt.savefig("plot.png")
plt.show()
```
