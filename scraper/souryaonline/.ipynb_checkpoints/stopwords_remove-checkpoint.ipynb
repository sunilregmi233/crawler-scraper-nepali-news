```python
from nltk.tokenize import word_tokenize
import nltk
import re

stopwords01 = open('../stopwords01.txt', 'r')
line = stopwords01.read()
words1 = line.split() 

stopwords = open('nepalnews_text.txt', 'r')
line1 = stopwords.read()
words2 = line1.split()
# print(words2)

word_tokens = word_tokenize(str(line1))

filtered_sentence = [w for w in word_tokens if not w in words1]
  
# filtered_sentence = [] 
  
# for w in word_tokens: 
#     if w not in stopwords01: 
#         filtered_sentence.append(w) 

print(filtered_sentence)

```

```python
# wordfreq = []
# for w in filtered_sentence:
#     wordfreq.append(filtered_sentence.count(w))
# print("Pairs\n" + str(list(zip(filtered_sentence, wordfreq))))

# mylist = list(set(wordfreq))
# mylist.sort() 
# term_freq = mylist[-20:]
# print(term_freq)

SYMBOLS = ',—।() ‘’'
results = []
for element in filtered_sentence:
    temp = ""
    for ch in element:
        if ch not in SYMBOLS:
            temp += ch

    results.append(temp)
    
while("" in results) : 
    results.remove("") 
#     results.remove("‘") 
    
print(results)

```

```python
# wordfreq = []
# for w in results:
#     wordfreq.append(results.count(w))
# pairs = str(list(zip(results, wordfreq)))
# # print("Pairs\n" + pairs)
# pairs[(1)]


from collections import Counter
list1=results
counts = Counter(list1)
print(counts)

twenty_words = dict(Counter(counts).most_common(20))
```
