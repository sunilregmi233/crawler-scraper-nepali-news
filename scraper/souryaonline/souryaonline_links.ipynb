```python
import requests
from bs4 import BeautifulSoup
from datetime import date, timedelta
import re

# with open('nepalnews_links.txt', "w") as f:

        
for i in range(1,50):
    with open('souryaonline_links.txt', "a+") as f:
        url = "https://www.souryaonline.com/category/khel-kheladi/page/" + str(i)
        getpage= requests.get(url)
        soup= BeautifulSoup(getpage.text, 'html.parser')
        ingredients = [ingredient for ingredient in soup.select('h2 a')]
        for j in ingredients:
            con = j['href']
            print(con)
            f.writelines(con + "\n")
```
