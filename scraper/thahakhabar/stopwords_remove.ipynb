```python
from nltk.tokenize import word_tokenize
import nltk
import re

stopwords01 = open('../stopwords01.txt', 'r')
line = stopwords01.read()
words1 = word_tokenize(str(line))


stopwords = open('thahakhabar_text.txt', 'r')
line1 = stopwords.read()
words2 = word_tokenize(str(line1))



SYMBOLS = ',—।()? ‘’१२३६४५७९०९०८–:'
results = []
for element in words2:
    temp = ""
    for ch in element:
        if ch not in SYMBOLS:
            temp += ch

    results.append(temp)
    
while("" in results) : 
    results.remove("") 
    



filtered_sentence = [w for w in results if not w in words1]
  


print(filtered_sentence)


with open("/home/sunil/Documents/Assignment_IS/crawler-scraper-nepali-news/5-sites-coupus-removed-stopwords.txt", "a+") as f:
    for text in filtered_sentence:  
        f.write("%s\n" % text)
    f.close()

```
